```python
import pandas as pd
import numpy as np
import csv
import re
```

```python
eye_data = pd.read_csv('log.txt',sep =r',(?!"y)',header = "infer", engine ='python')
```

```python
def createNestedDict (inputSentence, split) : 
    outdict = {}
    curdict = outdict
    for f in inputSentence.split(split):
        curdict[f] = {}
        curdict = curdict[f]
    return (outdict)
```

```python
finalDict = {}
#df = pd.DataFrame.from_records(nestedDictRow)
for i,line in enumerate(eye_data.iterrows()):
    categories = []
    catRegex  = re.compile(r'\"(.*?)\"')
    categories = re.findall(catRegex, str(line))
    
    values = []
    valRegex  = re.compile(r'(?<=:).*')
    values = re.findall(valRegex, str(line))
    rowDict = {}
    for cat,val in zip(categories,values):
        try:
            rowDict[str(cat)] = val
        except:
            rowDict[str(cat)] = None

    finalDict[i] = rowDict
```

```python
import time
def printDict (finalDict,wait_in_ms):
    for i in range(len(finalDict)):
        print(i,':')
        print(finalDict[i])
        print('========================')
        time.sleep(wait_in_ms*0.001)
```

```python
df = []
for i in range(len(finalDict)):
    df.append([])
    for j in range(len(finalDict[i])):
         df[i] = finalDict[i]
```

```python
columns = []
for row in df:
    valRegex  = re.compile(r'"(\w+)":')
    cat = re.findall(valRegex, row)
    columns.append(cat)
    time.sleep(30*0.001)
```

```python
df_columns = []
for cat in columns:
    df_columns.append(cat[0]) 
df_columns
```
